```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df=pd.read_csv("/content/drive/MyDrive/DataSets/Price_Agriculture_commodities_Week.csv")
df.head()
```

```python
#distribtion of data points by state
sns.countplot (data=df, x= 'State')
plt.xlabel('State')
plt.ylabel('Count')
plt.title('Distribution of data points by state')
plt.xticks(rotation=90)
plt.show()
```

```python
#distribution of data by districts
sns.set(font_scale= 0.5)
plt.figure(figsize= (30,10))
sns.countplot (data=df, x='District')
plt.xlabel('Commodity')
plt.ylabel('Count')
plt.title('Distribution of data by District')
plt.xticks(rotation= 90)
plt.show()
```

```python
#distribution of data by commodity
sns.set(font_scale= 0.5)
plt.figure(figsize= (20,10))
sns.countplot (data=df, x='Commodity')
plt.xlabel('Commodity')
plt.ylabel('Count')
plt.title('Distribution of data by Commodity')
plt.xticks(rotation= 90)
plt.show()
```

```python
comparePrice = df[['Min Price', 'Max Price', 'Modal Price']]
sns.boxplot(data=comparePrice)
plt.xlabel('Price Type')
plt.ylabel('Price')
plt.title('Price Distribution')
plt.show()
```

```python
#grade represent the quality of commodity
sns.countplot(x= 'Grade', data=df)
```

```python
#distribution of min price, max price and modal price through pie chart
from seaborn._core.properties import FontSize
labels='Min Price', 'Max Price', 'Modal Price'
size =[215,130, 245]
colors = ['gold', 'yellowgreen', 'lightcoral']
textprops={"fontsize":18}
explode= (0.1, 0, 0)
#plot
plt.pie(size, explode= explode, labels=labels, colors=colors, shadow= True, autopct='%1.1f%%', textprops=textprops )
plt.axis('equal')
plt.show()
```

```python
#hist plot of modal price and vs. frequency
plt.figure (figsize=(10,6))
sns.histplot(data=df, x='Modal Price', bins=25, color='skyblue')
plt.xlabel('Modal Price')
plt.ylabel('frequency')
plt.title('Modal price distribution')
plt.show()
```

```python
#scatter plot of max price and min price
plt.figure (figsize=(10,6))
sns.scatterplot(data=df, x='Max Price', y='Min Price', color='blue')
plt.xlabel('Max Price')
plt.ylabel('Min Price')
plt.title('Max Price and Min Price distribution')
plt.show()
```

```python
top_10_highest = df.nlargest(10, 'Modal Price')
top_10_lowest = df.nsmallest(10, 'Modal Price')
```

```python
top_10_highest.head(10)
```

```python
#using bar plot
plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
sns.barplot(data= top_10_highest, x='State', y='Modal Price')
plt.xlabel('State')
plt.ylabel('Modal Price')
plt.title(' State- wise Top 10 highest Modal price')
plt.show()
```

```python
#using bar plot
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.barplot(data= top_10_highest, x='Market', y='Modal Price')
plt.xlabel('Market')
plt.ylabel('Modal Price')
plt.title('Top 10 highest Modal price by market')
plt.xticks(rotation=90)
plt.show()
```

```python
#using bar plot
plt.figure(figsize=(18,6))
plt.subplot(1,2,1)
sns.barplot(data= top_10_highest, x='Commodity', y='Modal Price')
plt.xlabel('Commodity')
plt.ylabel('Modal Price')
plt.title('Top 10 highest Modal price by Commodity')
plt.xticks(rotation=90)
plt.show()
```

```python
#using bar plot
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.barplot(data= top_10_highest, x='Variety', y='Modal Price')
plt.xlabel('Variety')
plt.ylabel('Modal Price')
plt.title('Top 10 highest Modal price by Variety')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 2)
sns.barplot(data=top_10_lowest, x='Modal Price', y='Variety', hue='Commodity')
plt.xlabel('Modal Price')
plt.ylabel('Variety')
plt.title('Top 10 Lowest Prices by Variety')
```

```python
#using violin plot
#it help us to see both the distribution of data in terms of kernel density estimation and box plot)
plt.figure(figsize=(6,4))
sns.violinplot(data=df, y='Modal Price', palette= 'rainbow')
plt.ylabel('Modal Price')
plt.title('Modal Price Ditribution')
plt.tight_layout()
plt.show()
```

```python
price_columns= ['Min Price', 'Max Price', 'Modal Price']
plt.figure(figsize=(6,4))
sns.pairplot(data=df, vars= price_columns)
plt.tight_layout()
plt.show()
```

```python
price_columns=['Min Price', 'Max Price', 'Modal Price']
price_correlation= df[price_columns].corr()
plt.figure(figsize=(8,6))
sns.heatmap(price_correlation, annot= True, cmap= 'coolwarm', square= True)
plt.title('Price Correlation')
plt.tight_layout()
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Get the modal price for each commodity
modal_prices = df.groupby('Commodity')['Modal Price'].mean().sort_values(ascending=False)

# Create a dictionary of words and their frequencies
word_dict = {}
for commodity, price in modal_prices.items():
  word_dict[commodity] = price

# Create the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_dict)

# Plot the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Commodities by Modal Price')
plt.show()
```

```python
import chardet
with open('/content/drive/MyDrive/DataSets/city_locations.csv', 'rb') as f:
    encoding = chardet.detect(f.read())['encoding']
```

```python
df_loc = pd.read_csv("/content/drive/MyDrive/DataSets/city_locations.csv", encoding=encoding)
df_loc=pd.DataFrame(df_loc)
df_loc.head()
```

```python
df_loc=df_loc.rename(columns={'ASCII Name':'District'})
df_loc.head()
```

```python
df_loc=df_loc[['District','Latitude','Longitude']]
df_loc.head()
```

```python
merged_df = df.merge(df_loc, how='left', on='District')
merged_df.head()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load India map shapefile
india_map = gpd.read_file('/content/drive/MyDrive/DataSets/india_ds.shp')

# Assuming merged_df is your DataFrame containing the merged dataset with Latitude and Longitude columns
# Replace 'Latitude' and 'Longitude' with the actual column names if they are different
# Filter the DataFrame to get only rows with 'Tomato' as Commodity
tomato_df = merged_df[merged_df['Commodity'] == 'Tomato']

# Create a GeoDataFrame from the DataFrame with tomato prices
tomato_gdf = gpd.GeoDataFrame(tomato_df, geometry=gpd.points_from_xy(tomato_df['Longitude'], tomato_df['Latitude']))

# Plot India map
fig, ax = plt.subplots(figsize=(6, 6))
india_map.plot(ax=ax, color='#DCE8C0')

# Plot tomato prices
tomato_gdf.plot(ax=ax, color='red', markersize=5, alpha=0.5)

plt.title('Tomato Prices on India Map')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.tight_layout()
plt.show()
```
